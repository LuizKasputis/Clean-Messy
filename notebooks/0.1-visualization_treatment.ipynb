```python
import matplotlib.pyplot as plt
import matplotlib.image as implt
from PIL import Image 

import os
```

```python
train_data = '../data/raw/train/'
test_data = '../data/raw/test/'
val_data = '../data/raw/val/'

process_train = '../data/processed/train/'
process_test = '../data/processed/test/'
process_val = '../data/processed/val/'

category_names = ['messy/', 'clean/']
image_size = 50
```

```python
# Visualization Len messy and clean train

images_len_train = []

for category in category_names:
    
    dir_train = train_data + category
    len_category = len(os.listdir(dir_train))
    images_len_train.append(len_category)
    
fig = plt.figure()
ax = fig.add_axes([0,1,1,1])
ax.bar(category_names[0],images_len_train[0], color = 'b')
ax.bar(category_names[1],images_len_train[1], color = 'r')
plt.show()
```

```python
# Clean VS Messy imagens
Image.open(train_data+'/messy/0.png')
```

```python
# Clean VS Messy imagens
Image.open(train_data+'/clean/0.png')
```

```python
# GRAY_SCALE

messy_gray = Image.open(train_data+'/messy/0.png').convert('LA').resize((image_size, image_size), Image.ANTIALIAS)
clean_gray = Image.open(train_data+'/clean/0.png').convert('LA').resize((image_size, image_size), Image.ANTIALIAS)


plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1)
plt.imshow(clean_gray)
plt.axis('off')
plt.title("Clean in GrayScale")
plt.subplot(1, 2, 2)
plt.imshow(messy_gray)
plt.axis('off')
plt.title("Messy in GrayScale")
```

```python
# Train GrayScale with new size 50x50

for category in category_names:
    
    path = train_data+category
    path_processed = process_train+category
    
    for image in os.listdir(path):
        
        image_treatment = Image.open(path+image).convert('LA').resize((image_size, image_size), Image.ANTIALIAS)
        image_treatment.save(path_processed+image, 'PNG')
```

```python
# Test GrayScale with new size 50x50
for image in os.listdir(test_data):

    image_treatment = Image.open(test_data+image).convert('LA').resize((image_size, image_size), Image.ANTIALIAS)
    image_treatment.save(process_test+image, 'PNG')
```

```python
# Val GrayScale with new size 50x50

for category in category_names:
    
    path = val_data+category
    path_processed = process_val+category
    
    for image in os.listdir(path):
        
        image_treatment = Image.open(path+image).convert('LA').resize((image_size, image_size), Image.ANTIALIAS)
        image_treatment.save(path_processed+image, 'PNG')
```
