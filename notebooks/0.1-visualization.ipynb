```python
import matplotlib.pyplot as plt
import matplotlib.image as implt
from PIL import Image 

import os
```

```python
train_data = '../data/raw/train/'
test_data = '../data/raw/test/'
val_data = '../data/raw/val/'

process_train = '../data/processed/train/'
process_test = '../data/processed/test/'
process_val = '../data/processed/val/'

category_names = ['messy/', 'clean/']
image_size = 50
```

```python
# Visualization Len messy and clean train

images_len_train = []

for category in category_names:
    
    dir_train = train_data + category
    len_category = len(os.listdir(dir_train))
    images_len_train.append(len_category)
    
fig = plt.figure()
ax = fig.add_axes([0,1,1,1])
ax.bar(category_names[0],images_len_train[0], color = 'b')
ax.bar(category_names[1],images_len_train[1], color = 'r')
plt.show()
```

```python
# GRAY_SCALE with 50x50x2

messy_gray = Image.open(train_data+'/messy/0.png').convert('LA').resize((image_size, image_size), Image.ANTIALIAS)
clean_gray = Image.open(train_data+'/clean/0.png').convert('LA').resize((image_size, image_size), Image.ANTIALIAS)


plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1)
plt.imshow(clean_gray)
plt.axis('off')
plt.title("Clean in GrayScale")
plt.subplot(1, 2, 2)
plt.imshow(messy_gray)
plt.axis('off')
plt.title("Messy in GrayScale")
```

```python
# GRAY_SCALE with 224x224x3
image_size = 224

messy_gray = Image.open(train_data+'/messy/0.png').resize((image_size, image_size), Image.ANTIALIAS)
clean_gray = Image.open(train_data+'/clean/0.png').resize((image_size, image_size), Image.ANTIALIAS)


plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1)
plt.imshow(clean_gray)
plt.axis('off')
plt.title("Clean in RGB")
plt.subplot(1, 2, 2)
plt.imshow(messy_gray)
plt.axis('off')
plt.title("Messy in RGB")
```
